#### 2. Cálculo realizado: Correlação de Pearson ($r$)

O coeficiente de Pearson mede a correlação linear entre as variáveis.

**Somatórios obtidos ($n=27$):**
* $\sum X = 144,3$
* $\sum Y = 4.695$
* $\sum X^2 = 804,01$
* $\sum Y^2 = 1.050.425$
* $\sum XY = 27.696$

**Fórmula:**
$$r = \frac{n(\sum XY) - (\sum X)(\sum Y)}{\sqrt{[n\sum X^2 - (\sum X)^2][n\sum Y^2 - (\sum Y)^2]}}$$

**Substituição:**
* **Numerador:** $27(27.696) - (144,3)(4.695) = \mathbf{70.303,5}$
* **Denominador (X e Y):** $\sqrt{885,78 \cdot 6.318.450} \approx 74.811,47$

**Resultado:**
$$r \approx \frac{70.303,5}{74.811,47} \approx \mathbf{0,939}$$

```python
pearson_r, pearson_p = stats.pearsonr(df['Anos_Escolaridade'], df['Renda_Per_Capita'])

df_pearson = pd.DataFrame({
    'Coeficiente': ['Pearson (r)'],
    'Valor': [pearson_r],
    'P-Valor (Significância)': [pearson_p],
    'Conclusão': ['Significativo' if pearson_p < 0.05 else 'Não Significativo']
})

df_pearson
```

#### 3. Cálculo realizado: Correlação de Spearman ($\rho$)

O coeficiente de Spearman utiliza os postos (rankings). Considerando a soma dos quadrados das diferenças de postos ($\sum d^2 \approx 286$):

**Fórmula:**
$$\rho = 1 - \frac{6 \sum d^2}{n(n^2 - 1)}$$

**Substituição:**
$$\rho = 1 - \frac{6 \cdot 286}{27(728)} = 1 - \frac{1.716}{19.656} \approx 1 - 0,087$$

**Resultado:**
$$\mathbf{\rho \approx 0,913}$$

```python
spearman_r, spearman_p = stats.spearmanr(df['Anos_Escolaridade'], df['Renda_Per_Capita'])

df_spearman = pd.DataFrame({
    'Coeficiente': ['Spearman (ρ)'],
    'Valor': [spearman_r],
    'P-Valor (Significância)': [spearman_p],
    'Conclusão': ['Significativo' if spearman_p < 0.05 else 'Não Significativo']
})
df_spearman
```

#### 4. Teste de Significância

Considerando a Hipótese Nula ($H_0$) de que $r = 0$ com nível de significância de 5%.

* **Graus de liberdade:** $25$
* **$t_{crítico}$ (tabela):** $\approx 2,06$

**Cálculo do t obtido:**
$$t = r \sqrt{\frac{n-2}{1-r^2}} = 0,939 \sqrt{\frac{25}{1 - 0,881}} \approx 13,65$$

**Conclusão:**
Como $13,65 > 2,06$, a correlação é **estatisticamente significativa**.

```python
def highlight_sig(val):
    color = 'lightgreen' if val == 'Significativo' else 'white'
    return f'background-color: {color}'

print("\n--- Resultado Correlação de Spearman ---")
display(df_spearman.style.format({'Valor': '{:.4f}', 'P-Valor (Significância)': '{:.4e}'})
        .map(highlight_sig, subset=['Conclusão'])) 

print("--- Resultado Correlação de Pearson ---")
display(df_pearson.style.format({'Valor': '{:.4f}', 'P-Valor (Significância)': '{:.4e}'})
        .map(highlight_sig, subset=['Conclusão']))
```

Os cálculos confirmam uma correlação fortíssima e positiva (**$r \approx 0,94$** e **$\rho \approx 0,91$**) entre escolaridade e renda, corroborando estatisticamente os achados de *Salvato, Ferreira e Duarte (2010)*.

O artigo destaca que o diferencial de escolaridade explica entre **12% e 36%** das disparidades de renda regionais e que igualar a educação do Nordeste à do Sudeste elevaria a renda média daquela região em cerca de **55%**.
